```python
import pandas as pd


# Load the Excel file
df = pd.read_excel("geneexp_prediction.xlsx") 

# Define safe correlation function for both methods
def compute_correlations(group):
    x = group["DepMap_Expression"]
    y1 = group["Prediction Performance"]

    def safe_corr(x, y):
        if len(x) < 2 or x.std() == 0 or y.std() == 0:
            return "Not enough variance"
        return round(x.corr(y), 4)

    return pd.Series({
        "Correlation_Coefficient": safe_corr(x, y1),
    })

# Apply correlation per (Gene, MOA) group
correlation_df = (
    df.groupby(["Gene", "MOA"])
      .apply(compute_correlations)
      .reset_index()
)

# Merge correlation results back into the original dataframe
df_merged = df.merge(correlation_df, on=["Gene", "MOA"], how="left")
df_merged.to_excel("geneexp_prediction_corrcoeff.xlsx")
df_merged
```

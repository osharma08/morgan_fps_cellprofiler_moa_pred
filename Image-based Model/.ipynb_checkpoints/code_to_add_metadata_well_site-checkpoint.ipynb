```python
import pandas as pd
import glob
import os
import re
import string

# Function to convert row number to Excel-style letters (1=A, 2=B, ..., 26=Z, 27=AA, etc.)
def row_to_letter(row_num):
    letters = ""
    while row_num > 0:
        row_num, remainder = divmod(row_num - 1, 26)
        letters = chr(65 + remainder) + letters
    return letters

# Function to extract well and site from FileName_DNA
def extract_metadata(filename):
    match = re.match(r"r(\d{2})c(\d{2})f(\d{2})", filename)
    if match:
        row = int(match.group(1))
        col = int(match.group(2))
        site = int(match.group(3))

        well = f"{row_to_letter(row)}{col:02d}"
        return well, site
    return None, None

# Path to folder containing CSVs
folder_path = "/home/gauss/Documents/Project 2: Combining Molecular and Imagining Data for MOA Prediction/smile_image_global/Image-based Model/JUMP-Pilot Data/"

# Loop through all CSV files in the folder
for file in glob.glob(os.path.join(folder_path, "*.csv")):
    df = pd.read_csv(file)

    if "FileName_DNA" in df.columns:
        df["Metadata_Well"], df["Metadata_Site"] = zip(
            *df["FileName_DNA"].apply(extract_metadata)
        )

        # Save updated CSV (overwrite original or save with _updated suffix)
        output_file = os.path.splitext(file)[0] + "_updated.csv"
        df.to_csv(output_file, index=False)
        print(f"Processed and saved: {output_file}")
    else:
        print(f"Skipped (no FileName_DNA column): {file}")
```
